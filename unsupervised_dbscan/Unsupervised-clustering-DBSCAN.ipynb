# DBSCAN Clustering | Full Case-study(Wholesale Customers)

### Importing & Inspecting Data

```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
%matplotlib inline

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('wholesale-customers-data.csv')
df.head()
```

```python
df['Channel'].unique()
```

```python
df['Region'].unique()
```

# 

### Exploring Data

```python
sns.scatterplot(data=df, x=df['Milk'], y=df['Grocery'], hue='Channel', palette='coolwarm')
plt.show()
```

```python
droping_col = df.drop(['Channel', 'Region'], axis=1)
droping_col
```

```python
droping_col.corr()
```

```python
sns.clustermap(droping_col.corr(), annot=True)
```

#

### Preprocessing Data

```python
sc = StandardScaler()
```

```python
scaled_X = sc.fit_transform(df)
```

```python
scaled_X
```

#

### Model Building & Predicting

##### DBSCAN with Chosen Value of Epsilon

```python
model = DBSCAN(eps=2)
```

```python
model.fit(scaled_X)
```

```python
model.labels_
```

```python
sns.scatterplot(data=df, x=df['Milk'], y=df['Grocery'], hue=model.labels_, palette='rainbow')
plt.show()
```

```python
df['Clusters'] = model.labels_
df
```

```python
df[df['Clusters'] == -1]
```

```python
len(df[df['Clusters'] == -1])
```

#

##### DBSCAN with Unknown Value of Epsilon

```python
noise_percent = []

for i in np.linspace(0.001, 3, 50):

    dbscan = DBSCAN(eps=i)
    dbscan.fit(scaled_X)

    noise = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    noise_percent.append(noise)

noise_percent
```

```python
sns.lineplot(x=np.linspace(0.001, 3, 50), y=noise_percent)
```
