# K-Means Clustering | Full Case Study(Mall Customers)

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import  warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

## Importing Dataset

```python
mall = pd.read_csv('mall-customers.csv')
mall.head()
```

```python
mall.info()
```

## Extracting Features

```python
X = mall.iloc[:, 3:].values
X
```

## Exploring Data

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title('Relationship Between Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

## Model Training & Predicting

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
mall.head()
```

```python
mall_clusted = mall.copy()
```

```python
mall_clusted['clusters'] = y_kmeans
```

```python
mall_clusted
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.title('Customer Segmentation by Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

## Finding The Optimal Number of Clusters (Elbow | Knee Method)

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
wcss    
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

## Visualising Clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='k', s=200, label='Centroids')
plt.title('Customer Segmentation by Income and Spending Score with Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```
